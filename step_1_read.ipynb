```python
import pandas as pd

data = []

dataset = "datasets/nocola_ungrammatical_dev.txt"

with open(dataset, 'r') as file:
    for line in file:
        line = line.strip()
        sentence_parts = line.split('\t')
        if len(sentence_parts) == 2:
            sentence_text, class_label = sentence_parts
            data.append((sentence_text, class_label))

df = pd.DataFrame(data, columns=['text', 'label'])
df.head()
```

```python
df.label.unique()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already created the dataframe 'df' from the previous steps

# Count the occurrences of each class label
label_counts = df['label'].value_counts()

# Plot the distribution of class labels
plt.figure(figsize=(10, 6))
label_counts.plot(kind='bar')
plt.title('Distribution of Error Labels')
plt.xlabel('Error Label')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
